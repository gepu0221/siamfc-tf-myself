```python
import tensorflow as tf
import os
import scipy.io
import numpy as np



'''
(11, 11, 3, 96)
(5, 5, 48, 256)
(3, 3, 256, 384)
(3, 3, 192, 384)
(3, 3, 192, 32)
_conv_w_sz=np.array([11,5,3,3,3]) #the map size of filter(weight of conv net)
_conv_w_in_c=np.array([3,96,256,384,384])# the input channle number of filter
_conv_w_out=np.array([96,256,384,384,256])# the output number of feature map(the number of filter kernel)'''
_conv_w_sz=np.array([11,5,3,3,3]) #the map size of filter(weight of conv net)
_conv_w_in_c=np.array([3,48,256,192,192])# the input channle number of filter
_conv_w_out=np.array([96,256,384,384,32])# the output number of feature map(the number of filter kernel)
_layer_nums=5
#[_conv_w_sz[i],_conv_w_sz[i],_conv_w_in_c[i],_conv_w_out[i]]

def _import_from_matconvent(net_path):
    mat=scipy.io.loadmat(net_path)
    net_dot_mat=mat.get('net')
    #organize parameters to import
    params=net_dot_mat['params']
    params=params[0][0]
    params_names=params['name'][0]
    params_names_list=[params_names[p][0] for p in range(params_names.size)]
    params_values=params['value'][0]
    params_values_list=[params_values[p] for p in range(params_values.size)]
    return params_names_list,params_values_list

#find all parameters matching to codename(there should be only one)
def _find_params(x,params):
    matching=[s for s in params if x in s]
    assert len(matching)==1,('Ambiguous param name found')
    return matching

def save_para():
    #import the pretrain net param
    net_pre_floder='net_pre'
    net_file='baseline-conv5_e55.mat'
    net_path=os.path.join(net_pre_floder,net_file)
    params_names_list,params_values_list=_import_from_matconvent(net_path)
    
    for i in range(_layer_nums):
        conv_W_name=_find_params('conv'+str(i+1)+'f',params_names_list)[0]
        conv_b_name=_find_params('conv'+str(i+1)+'b',params_names_list)[0]
        conv_W=params_values_list[params_names_list.index(conv_W_name)]
        conv_b=params_values_list[params_names_list.index(conv_b_name)]
        
        print(conv_W.shape)
        conv_w_n='conv'+str(i+1)+'w'
        conv_b_n='conv'+str(i+1)+'b'
        w=tf.get_variable(conv_w_n,conv_W.shape,trainable=False,initializer=tf.constant_initializer(conv_W))
        b=tf.get_variable(conv_b_n,conv_b.shape,trainable=False,initializer=tf.constant_initializer(conv_b))
    init=tf.global_variables_initializer()
    #define the save param 'saver'
    saver=tf.train.Saver()
    
    with tf.Session() as sess:
        sess.run(init)
        #save the data in session
        save_path=saver.save(sess,'net_pre/save_net_param.ckpt')

def restore_para():
    for i in range(_layer_nums):
        conv_w_n='conv'+str(i+1)+'w'
        conv_b_n='conv'+str(i+1)+'b'
        w=tf.get_variable(conv_w_n,[_conv_w_sz[i],_conv_w_sz[i],_conv_w_in_c[i],_conv_w_out[i]],
                          trainable=False,initializer=tf.zeros_initializer())
        b=tf.get_variable(conv_b_n,[_conv_w_out[i],1],trainable=False,initializer=tf.zeros_initializer())
        
        
    saver=tf.train.Saver()
    
    with tf.Session() as sess:
        save_path=saver.restore(sess,'net_pre/save_net_param.ckpt')
        
restore_para()
```
